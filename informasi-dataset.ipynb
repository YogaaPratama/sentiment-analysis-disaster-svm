```python
import matplotlib.pyplot as plt

# Data
categories = ['Gempa Bumi Haiti 2010', 'Gempa Bumi Kroasia 2020', 'Banjir Catania 2021', 'Banjir Eropa Tengah 2021']
data_counts = [3238, 869, 207, 1120]

# Membuat diagram pie chart
plt.figure(figsize=(10,8))
plt.pie(data_counts, labels=categories, autopct='%1.1f%%', colors=['#F26B0F', '#E73879', '#FCC737', '#7E1891'], startangle=140)

# Menambahkan judul
plt.title('Distribusi Jumlah Data pada Dataset Bencana')

# Menampilkan diagram
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data total per kategori sentimen
labels = ["Negatif", "Netral", "Positif"]
values = [1654 + 8 + 86 + 64, 637 + 8 + 93 + 338, 600 + 2 + 17 + 88]  # Total dari semua wilayah
colors = {
    'positive': 'green',
    'negative': 'orange',
    'neutral': 'blue'
}

# Membuat Pie Chart
plt.figure(figsize=(7, 7))

# Menambahkan jumlah data ke label dan memetakan warna sesuai kategori
labels = [f"{label} ({value})" for label, value in zip(labels, values)]
mapped_colors = [colors['negative'], colors['neutral'], colors['positive']]

# Membuat diagram pie chart dengan label, warna, dan persentase
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=mapped_colors)

# Menambahkan judul pada diagram
plt.title("Distribusi Sentimen Secara Keseluruhan")

# Menampilkan diagram pie chart
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Data distribusi bahasa
catania_data = {
    'Italian': 126, 'English': 22, 'Swahili': 19, 'Spanish': 11, 'Portuguese': 4, 
    'Tagalog': 3, 'Arabic': 3, 'Catalan': 3, 'Turkish': 2, 'Dutch': 2, 
    'Polish': 1, 'Slovenian': 1, 'Greek': 1, 'French': 1, 'Hungarian': 1, 
    'Chinese (Traditional)': 1, 'German': 1, 'Albanian': 1
}

# Membuat word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(catania_data)

# Menampilkan word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Language Distribution in Dataset Catania")
plt.show()
```

```python
from wordcloud import WordCloud

# Data distribusi bahasa
croatia_data = {
    'Indonesian': 689, 'Finnish': 127, 'English': 17, 'Spanish': 14, 'Turkish': 11, 
    'Portuguese': 3, 'Latvian': 2, 'Croatian': 2, 'Italian': 2, 'French': 1
}

# Membuat word cloud
wordcloud = WordCloud(
    width=800, height=400, background_color='white', colormap='viridis',
    prefer_horizontal=1.0,  
    normalize_plurals=True, 
    min_font_size=5,  
    max_font_size=150  
).generate_from_frequencies(croatia_data)

# Menampilkan word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Language Distribution in Dataset Croatia")
plt.show()
```
